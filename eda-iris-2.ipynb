## Exploratatory Data Analysis.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
np.random.seed(123)
a=np.random.randint(0,101,50)
a
```

```python
sns.distplot(a,bins=10 ,kde=False ,hist_kws=dict(edgecolor="k",linewidth=2))
```

### Plotting Histogram.

```python
hist,bin_edges=np.histogram(a,bins=10)
print(hist,"\n\n",bin_edges)
```

```python
sns.distplot(a,bins=10  ,hist_kws=dict(edgecolor="k",linewidth=2))
```

```python
h,b=np.histogram(a,bins=10,density=True)
print(h,"\n\n",b)
```

### Calculating and Ploting Pdf ( Probability Density Function).

```python
pdf=h/sum(h)
pdf
```

```python
plt.plot(b[1:],pdf)
plt.ylabel("Probability")
```

```python
plt.plot(b[1:],pdf)
plt.xlabel("Probability")
```

### Importing the dataset.

```python
data=pd.read_csv("iris.csv")
data
```

### Scatter Plot.

```python
plt.scatter(data['sepal_length'],np.zeros(150))
```

```python
sns.distplot(data['sepal_length'])
plt.grid()
```

```python
sns.FacetGrid(data,hue='species',size=5).map(sns.distplot,'petal_length').add_legend()
plt.grid()
```

```python
sns.distplot(data['petal_width'])
plt.grid()
```

```python
sns.FacetGrid(data,hue='species',size=5).map(sns.distplot,'petal_width').add_legend()
plt.grid()
```

```python
sns.FacetGrid(data,hue='species',size=5).map(sns.distplot,'sepal_width').add_legend()
plt.grid()
```

```python
iris_setosa = data.loc[data['species'] == 'setosa']
iris_versicolor = data.loc[data['species'] == 'versicolor']
iris_virginica = data.loc[data['species'] == 'virginica']
```

```python
hist,bin_edges=np.histogram(iris_setosa['petal_length'],bins=10,density=True)
pdf=hist/(sum(hist))
cdf=np.cumsum(pdf)
plt.grid()
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.xlabel('PL_Setosa')
plt.ylabel('Probability')
```

```python
hist,bin_edges=np.histogram(iris_versicolor['petal_length'],bins=10,density=True)
pdf=hist/(sum(hist))
cdf=np.cumsum(pdf)
plt.grid()
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.xlabel('PL_Setosa')
plt.ylabel('Probability')
```

```python
hist,bin_edges=np.histogram(iris_virginica['petal_length'],bins=10,density=True)
pdf=hist/(sum(hist))
cdf=np.cumsum(pdf)
plt.grid()
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.xlabel('PL_Setosa')
plt.ylabel('Probability')
```

```python
print("Means of PW:")
print(np.mean(iris_setosa["petal_length"]))
print(np.mean(iris_versicolor["petal_length"]))
print(np.mean(iris_virginica["petal_length"]))
```
